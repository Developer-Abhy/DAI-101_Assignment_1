```python
def stats(df, cols):
    summary = []
    for col in cols:
        summary.append((col, 'mean', df[col].mean(), 'median', df[col].median(), 'std', df[col].std()))
    return summary
```

```python
print(stats(df, ['Cement', 'Strength', 'Water']))
```

## Conclusion


1. Higher cement content generally increases strength but shows diminishing returns.
2. Excess water negatively impacts concrete strength, confirming the water-cement ratio's importance.
3. Superplasticizers improve strength up to a point, after which additional amounts show minimal effect.
4. Strength gains rapidly in the first 28 days, then stabilizes, highlighting the curing period's importance.
5. Strength distribution is mostly normal, with a few high-strength outliers indicating special mix designs.
6. Outliers in cement suggest extreme mixes, while water outliers indicate unusually dry or wet mixes.
7. Superplasticizer usage is skewed, with some extreme values, indicating variability in mix designs.
8. Coarse and fine aggregates remain consistent, showing minimal variation across mixes.
